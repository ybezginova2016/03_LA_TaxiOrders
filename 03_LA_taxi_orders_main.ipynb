### 2.1 Stationarity

Stationarity is a key part of time series analysis. Simply put, stationarity means that the manner in which time series data changes is constant. A stationary time series will not have any trends or seasonal patterns. 

**We will use the Dickey Fuller test to check for stationarity in our data.** This test will generate critical values and a p-value, which will allow us to accept or reject the null hypothesis that there is no stationarity. If we reject the null hypothesis, that means we accept the alternative, which states that there is stationarity.

These values allow us to test the degree to which present values change with past values. If there is no stationarity in the data set, a change in present values will not cause a significant change in past values.

#### 2.1.1 Stationary statistics check using ***p-value*** 

```python
from statsmodels.tsa.stattools import adfuller

ts = data_resampled['num_orders']
st_test = adfuller(ts, regression='ctt')

print('If the value with index [1] is less than p-value = 0.05, then time series is stationary.')
print()

if st_test[1] < 0.05:
    print('Conclusion: Time series is stationary.')

else:

    print('Conclusion: Time series is NOT stationary.')
```

#### 2.1.2 Dickey-Fuller test

Временной ряд в таком виде не является стационарным, так как подвержен тренду, который меняется во времени. Это можно подтвердить критерием Дики-Фуллера. ***Нулевая гипотеза гласит: "Временной ряд является нестацинарным". Альтернативная: "Временной ряд стационарен".***

Next, let’s import the augmented Dickey-Fuller test from the statsmodels package. 

**Augmented Dickey-Fuller unit root test**: The Augmented Dickey-Fuller test can be used to test for a unit root in a univariate process in the presence of serial correlation. Documentation: https://www.statsmodels.org/dev/generated/statsmodels.tsa.stattools.adfuller.html

```python
from statsmodels.tsa.stattools import adfuller
```

Next, let’s pass our data frame into the adfuller method. Here, we specify the autolag parameter as “AIC,” which means that the lag is chosen to minimize the information criterion.

```python
adft = adfuller(data_resampled, autolag="AIC")
```

Next, let’s store our results in a data frame display it.

```python
output_data = pd.DataFrame({"Values":[adft[0],adft[1],adft[2],adft[3], adft[4]['1%'], adft[4]['5%'], adft[4]['10%']]  , "Metric":["Test Statistics","p-value","No. of lags used","Number of observations used", 
                                                        "critical value (1%)", "critical value (5%)", "critical value (10%)"]})
print(output_data)
```

**The null hypothesis of the Augmented Dickey-Fuller is that there is a unit root, with the alternative that there is no unit root. If the pvalue is above a critical size, then we cannot reject that there is a unit root.**

```python
p_value = adfuller(data['num_orders'])[1]
print('p-value =', p_value)
if p_value < 0.05:
    print('Null hypothesis is rejected, time series is stationary based on Dickey-Fuller test results.')
else:
    print('Null hypothesis is NOT rejected, time series is non-stationary based on Dickey-Fuller test results.')
```

### ***Conclusion:*** We can see that our data is not stationary from the fact that our p-value is greater than 5 percent and the test statistic is greater than the critical value. We can also draw these conclusions from inspecting the data, as we see a clear, increasing trend in the number of taxi orders.

### 2.2 Autocorrelation 
**Checking time series data for autocorrelation in Python is another important part of the analytic process. This is a measure of how correlated time series data is at a given point in time with past values, which has huge implications across many industries.**

***For example, if our taxi orders data has strong autocorrelation, we can assume that high numbers of taxi orders today suggest a strong likelihood that they will be high tomorrow as well.***

The Pandas data frame has an autocorrelation method that we can use to calculate the autocorrelation in our taxi orders data. Let’s do this for a one-month lag.

```python
# data autocorrelation with the previous hour
autocorrelation_lag1 = data_resampled['num_orders'].autocorr(lag=1)
print("1-Hour Lag: ", autocorrelation_lag1)
```

```python
# data autocorrelation with the previous hour
autocorrelation_lag2 = data_resampled['num_orders'].autocorr(lag=2)
print("2-Hours Lag: ", autocorrelation_lag2)
```

```python
# data autocorrelation with 3 previous hours
autocorrelation_lag3 = data_resampled['num_orders'].autocorr(lag=3)
print("3-Hours Lag: ", autocorrelation_lag3)
```

```python
# data autocorrelation with 6 previous hours
autocorrelation_lag6 = data_resampled['num_orders'].autocorr(lag=6)
print("6-Hours Lag: ", autocorrelation_lag6)
```

```python
# data autocorrelation with 9 previous hours
autocorrelation_lag9 = data_resampled['num_orders'].autocorr(lag=9)
print("9-Hours Lag: ", autocorrelation_lag9)
```

```python
# data autocorrelation with 24 previous hours (1 day)
autocorrelation_lag24 = data_resampled['num_orders'].autocorr(lag=24)
print("24-Hours Lag: ", autocorrelation_lag24)
```

```python
# data autocorrelation with 168 previous hours (7 days)
autocorrelation_lag168 = data_resampled['num_orders'].autocorr(lag=168)
print("1-Week Lag: ", autocorrelation_lag168)
```

### 3.1 Creating new features

```python
def make_features(data, column, max_lag, rolling_mean_size):
    # creating new features
#     data['year'] = data.index.year
#     data['month'] = data.index.month
#     data['day'] = data.index.day
    data['hour'] = data.index.hour
    data['dayofweek'] = data.index.dayofweek
    
    # https://www.mikulskibartosz.name/forecasting-time-series-using-lag-features/
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data[column].shift(lag)
        
    # опрделение ширины окна для скользящего среднего
    data['rolling_sum'] = data[column].shift().rolling(rolling_mean_size).sum()

    return data
```

<div class="alert alert-warning">
<b>Importance of Hyperparameters</b>    

    
Hyperparameters ***max_lag, rolling_mean_size*** are really important for model training. Their values can not be too small, since small values of lagged features and small value of window size will not allow to traint the model properly or lead to overfitting.
    
- ***lag***
    
Creating lagged features for future modelling, we basically build an autocorrelation on purpose. And the future ML will take into account newly-created lagged values for prediction.
    
- ***rolling_mean_size***
    
For example, ***moving average (rolling mean)*** is a pure part of a trend. However, if we take a window size = 5 hours, then a daily seasonality, which exists in taxi orders time series, will create a lot of noise in the data. Hence, it is better to take a window size = 24 hours (1 day) in order to escape the weekly seasonality.

</div>

<b>Дизайн ML исследования</b>

В курсе мы используем следующую схему оценки качества алгоритмов:

1. Все данные делятся на 3 части: 
   - обучение,
   - валидация,
   - тест.

2. Каждый из алгоритмов будет обучен на обучающей выборке. 
3. Подбор гиперпараметров мы осуществляем с помощью оценки на валидационной выборке.
4. Когда гиперпараметры подобраны, качество каждого алгоритма мы оцениваем с помощью валидационной выборки.
    
5. Выбираем лучший по оценке на валидационной выборке.
6. И уже самый лучший алгоритм оцениваем по тестовой выборке.


Есть более системный вариант этой схемы, с использованием кросс-валидации. В этом случае  схема немного адаптируется. 

Все данные делятся на 2 части, первая будет использоваться для обучения и подбора гиперпараметров, и оценки алгоритмов с помощью кросс-валидации, вторая будет использоваться для финального тестирования.
